## NumPy: Array Loading, Saving, and Basic Data Transformations

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg  # It is just for loading the image
```

#### NumPy Array

```python
array_1 = np.array([[1,2,3], [4,5,6]])
array_2 = np.random.rand(3,3)
array_3 = np.zeros((4,4))
```

#### Save the Array to Flat Files

```python
np.save("array_1.npy", array_1)
np.save("array_2.npy", array_2)
np.save("array_3.npy", array_3)
```

#### Load the Saved Array

```python
loaded_a1 = np.load("array_1.npy")
print(loaded_a1)
```

#### Display a .npy image and generate its dark mode using Matplotlib

```python
# Load and Display the .npy file
try:
    logo = np.load("Numpy_Logo.npy")

    # Display .npy array image by using matplotlib
    plt.figure(figsize = (10, 5))
    plt.subplot(121) # create 1 row, 2 columns, and use the 1st subplot area
    plt.imshow(logo)
    plt.title("Numpy Logo")
    plt.grid(False)

    # Dark Mode

    # take only RGB channels (remove alpha channel A)
    # shape changes from (H, W, 4) → (H, W, 3)
    logo_rgb = logo[:, :, :3]

    # invert colors for dark mode
    # since values are 0–1: 0 becomes 1 (black→white), 1 becomes 0 (white→black)
    dark_logo = 1 - logo_rgb

    plt.subplot(122) # create 1 row, 2 columns, and use the 2nd subplot area
    plt.imshow(dark_logo)
    plt.title("Numpy Dark Logo")
    plt.grid(False)
    
except FileNotFoundError:
    print("Numpy_Logo.npy file not found")
```
