## Advance Operation with Business Examples

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data Structure: [Resturent_Id, 2021, 2022, 2023, 2024]
sales_data = np.array([
    [1, 150000, 180000, 220000, 250000], # Biryani Point
    [2, 120000, 140000, 160000, 190000], # Beiging Bites
    [3, 200000, 230000, 260000, 300000], # Pizza Max
    [4, 180000, 210000, 240000, 270000], # Burger Point
    [5, 160000, 180000, 205000, 230000]  # Chai Point
])

print("=======>> Foodpanda Sales Analysis <<=======")
print("Data Size: ", sales_data.shape)
print("\nSales of first three restaurants:\n", sales_data[:3]) # [0:3]
print("\nSales of all restaurants without ID:\n", sales_data[:, 1 :]) # skip first column ---> [rows, column[start : stop]]
print("\nSales of all restaurants without the first row::\n", sales_data[1: ,:]) # skip first row ---> [rows[start : stop] , column]
```

```python
# Total sales per year
print("The total sale per year is: ",np.sum(sales_data, axis=0)) # calculation the total sales per year
print("Sales data without id: ", np.sum(sales_data[:, 1:], axis=0)) # calculating the total sales without calculating the id

# Minimum slales per restirent

print("Minimum sale per resturent: ", np.min(sales_data[:, 1:], axis=1)) # minimum sales (axis=1 means left to right)

# Maximum sales per year
print("Maximum sale per year: ", np.max(sales_data[:, 1:], axis=0))

# Average sales per resturent
print("Average sales per resturent: ", np.mean(sales_data[:, 1:], axis=1)) # calculate the average sale per resturent
```

```python
# Cumulative sum of restaurants for each year
cumsum = np.cumsum(sales_data[:, 1:], axis=1)
print("Cumulative Sum of the restaurants:\n", cumsum) # cumsum (cumulative sum): adds values year by year and gives the running total

# Matplotlib plotting 
plt.figure(figsize=(8,5))
plt.plot(np.mean(cumsum, axis=0))
plt.title("Average cummulative sales across all resturent")
plt.xlabel("Years")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```

```python
vector_1 = np.array([1, 2, 3, 4, 5])  # Create first vector
vector_2 = np.array([6, 7, 8, 9, 10])  # Create second vector

# Adds corresponding elements of both vectors (element-wise addition)
print("Vector Addition: ", vector_1 + vector_2)  

# Multiplies corresponding elements (element-wise multiplication, NOT dot product)
print("Vector Multiplication: ", vector_1 * vector_2)  

# Computes dot product (sum of element-wise multiplications)
print("Dot product: ", np.dot(vector_1, vector_2))  

# Calculates angle between vectors using cosine formula
angle = np.arccos(
    np.dot(vector_1, vector_2) / 
    (np.linalg.norm(vector_1) * np.linalg.norm(vector_2)) # norm means normalization which is also called modelous
)  

print(angle)  
# Prints the angle (in radians)
```

```python
# Vectorize Operation ---> Applies a function to each element of an array without writing explicit loops

restaurent_type = np.array(["Biryani", "chai", "pizza", "burger", "cafe"])  
# Array containing string values

vectorize_upper = np.vectorize(str.upper)  
# Converts a normal python function into a vectorized function to apply it element-wise on an array

# Applies the function to each element and prints the transformed array
print("Vectorize Upper: ", vectorize_upper(restaurent_type))  
```

```python
# Broadcasting ---> NumPy automatically performs operations between arrays and scalars (or compatible shapes)
# It expands dimensions internally and applies operations element-wise without explicit loops (vectorize)
# Best used for numeric operations like +, -, *, / because it is fast and optimized at C-level

# np.vectorize ---> Wraps a Python function (built-in like str.upper OR custom functions) so it can be applied element-wise on arrays
# It works like a loop internally, so it is mainly for convenience, not performance
# It does NOT improve speed compared to broadcasting or true NumPy operations

# Divides each element by 12 using broadcasting (fast and automatic)
monthly_avg = sales_data[:, 1:] / 12  

print(monthly_avg)
```
